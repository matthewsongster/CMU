```python
counties = pd.read_csv('coord2_clean.csv')
clusters = pd.read_csv('cluster_assignment.csv')
data = pd.merge(counties, clusters, left_index=True, right_index=True)
data.head(5)
```

```python
area = []
for i in data.index:
  county = data['Recip_County'][i]
  state = data['Recip_State'][i]
  area.append(county+', '+state)
data['Name'] = area
data.head(5)
```

```python
covid = pd.read_csv('covid_clean.csv')
covid = covid.loc[covid['Year'] == 2021]
area_2 = []
for i in covid.index:
  county = covid['Recip_County'][i]
  state = covid['Recip_State'][i]
  area_2.append(county+', '+state)
covid['Name'] = area_2
covid.head(5)
```

```python
def get_cluster(name):
  return list(data.loc[data['Name'] == name]['x'])[0]
cluster_num = []
for i in covid.index:
  name = covid['Name'][i]
  cluster_num.append(get_cluster(name))
covid['Cluster Number'] = cluster_num
covid.head(5)
```

```python
def demand(cluster_df, d):
  return cluster_df[d]
```

```python
I = 10000 #our fixed costs assume this amount of inventory, so we would not be able to store more than this
per_batch = 7114
#Cost function 
#Main Assumptions:
#Filling entire capacity maximizes the cost (If we make d items)
#Making twice the entire capacity minimizes the cost as we don't have to make anything the following week.
#modify
def c(x, d): 
    fixed = 4333.33
    c1 = -1 / d
    return c1 * (x-d)**2 + (d + fixed)
```

```python
#Backward Induction -- def will need to be debugged ! 
def inventory_planning(matrix, df):
  for i in range(I): #base case for final time period
    produce = demand(df, 12) - i*7114
    matrix[11][i] = max(c(produce, I),0)
  for t in range(10, -1, -1):
    for i in range(I):
      units = demand(df, t+1) - i*7114 #modify --> demand should be pulled from final table
      l = [0 for x in range(I)]
      for inv in range(len(l)):
        cost = c((units + inv*7114), I) + matrix[t+1][inv]
        l[inv]= cost
      matrix[t][i] = min(l)
  return matrix
```
